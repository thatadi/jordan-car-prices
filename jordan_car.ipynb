```python
import pandas as pd
import numpy as np
import tensorflow as tf
from tensorflow import keras
from keras.models import Sequential
from keras.layers import Dense,Dropout
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import MinMaxScaler
from sklearn.linear_model import LinearRegression
from sklearn.preprocessing import LabelEncoder
from sklearn.metrics import r2_score,mean_squared_error
from sklearn.preprocessing import StandardScaler

```

```python
df=pd.read_csv('car_prices_jordan.csv')
df
```

```python
df['Price'] = df['Price'].str.replace(',', '').astype(float)
```

```python
df.describe()
```

```python
df.info()
```

```python
df.drop(df.columns[0], axis=1, inplace=True)
```

```python
X=df.drop('Price', axis=1)
x=pd.get_dummies(X)
y=df['Price']
```

```python
train_x,test_x,train_y,test_y=train_test_split(x,y,test_size=0.3,random_state=42)
```

```python
scaler = StandardScaler()
X_train = scaler.fit_transform(train_x)
X_test = scaler.transform(test_x)
```

```python
model= Sequential()
model.add(Dense(units = 20, activation = 'relu',input_shape=(train_x.shape[1],)))
model.add(Dense(units = 10, activation = 'relu'))
model.add(Dense(units = 4, activation = 'relu'))
model.add(Dense(units = 1,activation='linear'))
```

```python
model.compile(optimizer='adam', loss='mean_squared_error', metrics=['mae'])
model.fit(train_x, train_y, batch_size=1,  epochs = 100)
```

```python
pred_y=model.predict(test_x)
```

```python
print("R2_Score : " ,r2_score(test_y,pred_y))
print("Mean Squared Error : ", mean_squared_error(test_y,pred_y))
print("Root Mean Squared Error : ", np.sqrt(mean_squared_error(test_y,pred_y)))
```
